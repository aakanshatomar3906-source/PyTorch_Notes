REDUCTION OPERATION

```python
e=torch.randint(size=(2,3),low=0,high=10)
e
```

```python
# sum
torch.sum(e)

# sum along columns
torch.sum(e,0)
torch.sum(e,dim=0)

# sum along rows
torch.sum(e,dim=1)
```

```python
# mean
#torch.mean(e)  # error=> only work with float point number.
f=torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
torch.mean(f)

# mean along column
torch.mean(f,dim=0)
```

```python
# median
torch.median(f)
```

```python
# max and min
torch.max(f)
torch.min(f)
```

```python
# product
torch.prod(e)
```

```python
# standard deviation
torch.std(f)
```

```python
# variance
torch.var(f)
```

```python
# argmax
torch.argmax(e)
```

```python
# argmin
torch.argmin(e)
```

MATRIX OPERATIONS:-

```python
m=torch.randint(size=(2,3),low=0,high=10)
n=torch.randint(size=(3,2),low=0,high=10)
print(m)
print(n)
```

```python
# matrix multiplication
torch.matmul(m,n)
```

```python
vector1=torch.tensor([1,2])
vector2=torch.tensor([3,4])

# dot product
torch.dot(vector1,vector2)
```

```python
# transpose
torch.transpose(m,0,1)
```

```python
h=torch.randint(size=(3,3),low=0,high=10,dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

5. COMPARISON OPERATIONS

```python
i=torch.randint(size=(2,3),low=0,high=10)
j=torch.randint(size=(2,3),low=0,high=10)

print(i)
print(j)
```

```python
# greater than
i>j

# less than
i<j

# equal to
i==j

# not equal to
i!=j

# greater than equal to
i>=j
```
